```python

df = ['Australia_Wine_Stats.csv','Chile_Wine_Stats.csv','France_Wine_Stats.csv','Italy_Wine_Stats.csv','New Zealand_Wine_Stats.csv','Portugal_Wine_Stats.csv','Spain_Wine_Stats.csv','USA_Wine_Stats.csv']
df
```

```python
wine_df = pd.concat((pd.read_csv(i) for i in df)).reset_index(drop=True)
wine_df
```

```python
wine_df = wine_df.drop_duplicates()
```

```python
wine_df = wine_df.dropna()
wine_df['Country'] = wine_df['Region']
wine_df['Country'] = wine_df['Region'].str.split('/').str[0].str.strip()
wine_df
```

```python
import dash
import plotly.express as px
import pandas as pd
from dash import dcc, html
from dash.dependencies import Input, Output
import os



# Initialize the Dash app
app = dash.Dash(__name__)

# Define the visualizations (charts) initially
# Bar Chart: Wine Prices by Country/Region

bar_chart = px.bar(wine_df, x='Price', y='Country', color='Country', title="Wine Prices by Region")

# Scatter Plot: Rating vs. Price
scatter_plot = px.scatter(wine_df, x='Rating', y='Price', color='Wine style', title="Price vs. Rating")

box_plot = px.box(wine_df, y='Alcohol content')
 
# Pie Chart: Wine Style Distribution
pie_chart = px.pie(wine_df, names='Wine style', title="Wine Style Distribution")

# Histogram: Alcohol Content Distribution
hist_chart = px.histogram(wine_df, x='Price', nbins=30, title="Price Content Distribution")

# Heatmap: Correlation between Wine Attributes
correlation = wine_df[['Price', 'Rating', 'Alcohol content', 'Bold', 'Tannin', 'Sweet', 'Acidic']].corr()
heatmap_chart = px.imshow(correlation, title="Wine Attribute Correlations")

# Layout of the dashboard
app.layout = html.Div([
    html.H1("Wine Data Dashboard", style={'text-align': 'center'}),

    # Dropdown for selecting wine region
    dcc.Dropdown(
        id='Country-dropdown',
        options=[{'label': Country , 'value':Country } for Country in wine_df['Country'].unique()],
        value='Australia',  # Default value
        multi=False,
        style={'width': '50%', 'padding': '20px'}
    ),

    # Charts
     dcc.Graph(id='bar-chart', figure=bar_chart),
    dcc.Graph(id='box-chart', figure=box_plot),
     dcc.Graph(id='scatter-chart', figure=scatter_plot),
     dcc.Graph(id='pie-chart', figure=pie_chart),
    dcc.Graph(id='hist-chart', figure=hist_chart),
    dcc.Graph(id='heatmap-chart', figure=heatmap_chart)
])

# Callback to update charts based on dropdown selection
@app.callback(
    [Output('bar-chart', 'figure'),
     Output('box-chart', 'figure'),
     Output('scatter-chart', 'figure'),
     Output('pie-chart', 'figure'),
     Output('hist-chart', 'figure'),
     Output('heatmap-chart', 'figure')],
    [Input('Country-dropdown', 'value')]
)
def update_charts(selected_region):
    filtered_df = wine_df[wine_df['Country'] == selected_region]

    # Update figures
    bar_fig = px.bar(filtered_df, x='Region', y='Price', color='Region', title=f"Wine Prices in {selected_region}")
    box_fig = px.box(filtered_df, y='Alcohol content', color='Region', title=f"Wine Prices in {selected_region}")
    scatter_fig = px.scatter(filtered_df, x='Price', y='Rating', color='Wine style', title=f"Price vs. Rating in {selected_region}")
    pie_fig = px.pie(filtered_df, names='Wine style', title=f"Wine Style Distribution in {selected_region}")
    hist_fig = px.histogram(filtered_df, x='Alcohol content', nbins=30, title=f"Alcohol Content in {selected_region}")
    corr = filtered_df[['Price', 'Rating', 'Alcohol content', 'Bold', 'Tannin', 'Sweet', 'Acidic']].corr()
    heatmap_fig = px.imshow(corr, title=f"Wine Attribute Correlations in {selected_region}")
    
    return bar_fig, scatter_fig, pie_fig, hist_fig, heatmap_fig

# Run the app
if __name__ == '__main__':
    app.run_server(debug=True)
```
